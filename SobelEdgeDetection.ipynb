```
import cv2
import numpy as np
from matplotlib import pyplot as plt
import math   
```

```
filter_x = [[1,0,-1], [2,0,-2], [1,0,-1]]
filter_y = [[1,2,1], [0,0,0], [-1,-2,-1]]
def sobelDetect(img):
  new_img=img.copy()
  height=img.shape[0]
  width=img.shape[1]
  for x in range(3, height - 2):
    for y in range(3, width - 2):
      # 3x3 matrix
      matrix = img[x-1:x+2,y-1:y+2,0]
      dy_temp = filter_y*matrix
      dy=dy_temp.sum() / 4

      dx_temp = filter_x * matrix
      dx = dx_temp.sum() / 4
      val = math.sqrt(dx*dx + dy * dy)
      new_img[x,y] = val * 3
   
  return new_img

def showOutput(original,result,originalTitle="original",result_title="Result"):
  plt.figure(figsize=(18,12))
  plt.subplot(121)
  plt.imshow(original)
  plt.title(originalTitle)
  plt.xticks([])
  plt.yticks([])
  plt.subplot(122)
  plt.imshow(result)
  plt.title(result_title)
  plt.xticks([])
  plt.yticks([])
  plt.show()

img = plt.imread('car.jpg')
img2=plt.imread("cat.jpg")  
 
img_binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1] 
new_img=sobelDetect(img) 
showOutput(img,new_img,"Original","result")
new_img=sobelDetect(img_binary)
showOutput(img_binary,new_img,"grayed Image","result")

new_img=sobelDetect(img2)
showOutput(img2,new_img,"Original","result")

img_binary2 = cv2.threshold(img2, 127, 255, cv2.THRESH_BINARY)[1]
new_img=sobelDetect(img_binary2)
showOutput(img_binary2,new_img,"grayed Image","result")
```
